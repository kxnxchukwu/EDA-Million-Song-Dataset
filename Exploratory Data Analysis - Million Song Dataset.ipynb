## Data Cleaning

### Checking Null Values in the Dataset

```python
MSD.isnull().sum()
```

### Drop Columns with Zeros all through

```python
MSD = MSD.drop(columns=['Unnamed: 0', 'danceability', 'energy', 'idx_bars_confidence', 'idx_bars_start', 'idx_beats_confidence', 'idx_beats_start', 'idx_artist_terms', 'idx_similar_artists', 'idx_artist_mbtags', 'idx_sections_confidence', 
                        'idx_sections_start', 'idx_segments_confidence', 'idx_segments_loudness_max', 'idx_segments_loudness_max_time', 'idx_segments_loudness_start', 'idx_segments_pitches', 'idx_segments_start', 'idx_segments_timbre',
                       'idx_tatums_confidence', 'idx_tatums_start', 'analyzer_version'])
```

```python
MSD.head(5)
```

### Drop Rows where Year equal Zero

```python
MSD = MSD.drop(MSD[MSD.year == 0].index)
```

```python
len(MSD)
```

```python
MSD.head(5)
```

### Dropping NAs values

```python
#MSD = MSD.dropna() This dropped all the values in the dataframe.
```

### Check dataframe for Duplicate Records based on Song ID

```python
duplicateSongs = MSD[MSD.duplicated(['song_id'])]
print("Duplicate Rows based on Song ID column are:", duplicateSongs, sep='\n')
```

### Drop Duplicate Records based on Song ID

```python
MSD = MSD.drop_duplicates(subset=['song_id'], keep='last')
```

```python
len(MSD)
```

### Check dataframe for Duplicate Records based on 7 Digital Track ID

```python
duplicateDigitalID = MSD[MSD.duplicated(['track_7digitalid'])]
print("Duplicate Rows based on 7 Digital ID column are:", duplicateDigitalID, sep='\n')
```

### Drop Duplicate Records based on 7 Digital Track ID

```python
MSD = MSD.drop_duplicates(subset=['track_7digitalid'], keep='last')
```

```python
len(MSD)
```

```python
MSD.isnull().sum()
```

## Exploratory Data Analysis

#### Arstists with the Most Number of Songs in the Dataset

```python
mostSongsbyArtist_df = MSD['artist_name'].value_counts().rename_axis('Artist').to_frame('counts')

print(mostSongsbyArtist_df.head(10))
```

#### Songs Titles with the Most Numbers in the Dataset

```python
songs_df = MSD['title'].value_counts().rename_axis('Song Title').to_frame('counts')

print(songs_df.head(10))
```

### Number of Song(s) per Year

```python
yearStatus = pd.value_counts(MSD['year'].values, sort=True)
print(yearStatus)
```

```python
yearStatus.plot.barh(figsize=(20,15))
plt.title('Horizontal Bar Plot of the Number of Song(s) per Year', fontdict=None, loc='center', pad=None)
```

### Mean Song Hotness by Year

```python
MSD.groupby('year')['song_hotttnesss'].mean()
```

```python
plt.figure(figsize=(20,10))
plt.plot(MSD.groupby('year')['song_hotttnesss'].mean())
plt.title('Line Plot of the Mean Song Hotness by Year', fontdict=None, loc='center', pad=None)
```

### Mean Tempo by Year

```python
MSD.groupby('year')['tempo'].mean()
```

```python
plt.figure(figsize=(20,10))
plt.plot(MSD.groupby('year')['tempo'].mean())
plt.title('Line Plot of the Mean Tempo by Year', fontdict=None, loc='center', pad=None)
```

### Mean Loudness by Year

```python
MSD.groupby('year')['loudness'].mean()
```

#### Scatter Plot of Year and Duration

```python
MSD.plot.scatter(x="year", y="duration")
plt.title('Scatter Plot of the Year and Duration', fontdict=None, loc='center', pad=None)
```

#### Correlation HeatMap

```python
corr = MSD[['song_id', 'key', 'mode', 'duration', 'loudness', 'tempo', 'song_hotttnesss', 'year']].corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

#### Pie Chart Showing Genres in the Dataset

```python
genres = pd.value_counts(MSD['genre'].values, sort=True)
genreLabel= MSD.genre.unique()
```

```python
plt.figure(figsize=(45,25))
genres.plot.pie()
plt.legend(genreLabel, loc="best")
plt.title('Pie Chart Showing the Genres in the Dataset', fontdict=None, loc='center', pad=None)
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig = plt.figure(figsize=(25,10))

m = Basemap(projection='mill',
           llcrnrlat = -90,
           urcrnrlat = 90,
           llcrnrlon = -180,
           urcrnrlon = 180,
           resolution = 'c')

m.drawcoastlines()

#Fill the globe with a blue color 
m.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
m.fillcontinents(color='coral',lake_color='aqua')

m.drawparallels(np.arange(-90,90,10),labels=[True,False,False,False])
m.drawmeridians(np.arange(-180,180,30),labels=[0,0,0,1])

sites_lat_y = MSD['artist_latitude'].tolist()
sites_lon_x = MSD['artist_longitude'].tolist()

m.scatter(sites_lon_x,sites_lat_y,latlon=True, s=500, marker='o', alpha=1, edgecolor='k', linewidth=1, zorder=2)

plt.title('Artiste\'s Locations', fontsize=20)

plt.show()
```
